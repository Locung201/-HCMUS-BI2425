```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Read and prepare the data
def load_and_prepare_data(data):
    # Convert data to DataFrame
    df = pd.read_csv(data)
    
    # Clean column names
    df.columns = df.columns.str.strip()
    
    # Convert Date column to datetime
    df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
    
    # Convert AQI to numeric
    df['AQI'] = pd.to_numeric(df['AQI'])
    
    return df

def profile_data(df):
    """
    Generate a comprehensive profile of the air quality data
    """
    profile = {
        'basic_stats': {
            'total_records': len(df),
            'date_range': f"{df['Date'].min()} to {df['Date'].max()}",
            'unique_counties': df['county Name'].nunique(),
            'unique_categories': df['Category'].unique().tolist(),
            'aqi_stats': df['AQI'].describe().to_dict()
        },
        'quality_metrics': {
            'missing_values': df.isnull().sum().to_dict(),
            'duplicate_records': df.duplicated().sum()
        }
    }
    
    # Calculate category distribution
    category_dist = df['Category'].value_counts().to_dict()
    profile['category_distribution'] = category_dist
    
    # Calculate monthly averages
    df['Month'] = df['Date'].dt.month
    monthly_avg = df.groupby('Month')['AQI'].mean().to_dict()
    profile['monthly_averages'] = monthly_avg
    
    # Calculate county-wise statistics
    county_stats = df.groupby('county Name')['AQI'].agg(['mean', 'min', 'max']).to_dict()
    profile['county_statistics'] = county_stats
    
    return profile

def plot_aqi_trends(df):
    """
    Create visualizations for AQI trends
    """
    plt.figure(figsize=(15, 10))
    
    # Create subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 15))
    
    # 1. Time series plot
    sns.lineplot(data=df, x='Date', y='AQI', ax=ax1)
    ax1.set_title('AQI Trends Over Time')
    ax1.tick_params(axis='x', rotation=45)
    
    # 2. Box plot of AQI by Category
    sns.boxplot(data=df, x='Category', y='AQI', ax=ax2)
    ax2.set_title('AQI Distribution by Category')
    ax2.tick_params(axis='x', rotation=45)
    
    # 3. Monthly AQI distribution
    df['Month'] = df['Date'].dt.month
    sns.boxplot(data=df, x='Month', y='AQI', ax=ax3)
    ax3.set_title('Monthly AQI Distribution')
    
    # 4. AQI distribution by county
    sns.boxplot(data=df, x='county Name', y='AQI', ax=ax4)
    ax4.set_title('AQI Distribution by County')
    ax4.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    return fig

# Function to generate summary report
def generate_summary_report(profile):
    """
    Generate a formatted summary report from the profile data
    """
    report = """
    Air Quality Data Analysis Report
    ==============================
    
    Basic Statistics:
    ----------------
    Total Records: {total_records}
    Date Range: {date_range}
    Number of Counties: {unique_counties}
    
    AQI Statistics:
    --------------
    Mean AQI: {mean_aqi:.2f}
    Median AQI: {median_aqi:.2f}
    Min AQI: {min_aqi:.2f}
    Max AQI: {max_aqi:.2f}
    
    Category Distribution:
    -------------------
    {category_dist}
    
    Data Quality:
    -----------
    Missing Values: {missing_values}
    Duplicate Records: {duplicates}
    """.format(
        total_records=profile['basic_stats']['total_records'],
        date_range=profile['basic_stats']['date_range'],
        unique_counties=profile['basic_stats']['unique_counties'],
        mean_aqi=profile['basic_stats']['aqi_stats']['mean'],
        median_aqi=profile['basic_stats']['aqi_stats']['50%'],
        min_aqi=profile['basic_stats']['aqi_stats']['min'],
        max_aqi=profile['basic_stats']['aqi_stats']['max'],
        category_dist='\n    '.join([f"{k}: {v}" for k, v in profile['category_distribution'].items()]),
        missing_values=sum(profile['quality_metrics']['missing_values'].values()),
        duplicates=profile['quality_metrics']['duplicate_records']
    )
    
    return report
```

## State AQI 2021

```python
df = load_and_prepare_data('Data Cleaned/10_state_aqi_2021_cleaned.csv')
profile = profile_data(df)
print(generate_summary_report(profile))
fig = plot_aqi_trends(df)
plt.show()
```
